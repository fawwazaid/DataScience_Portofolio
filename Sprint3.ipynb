````python
# Pengecekan nilai yang hilang pada kolom model_year karena kemungkinan besar tipe yang sama memiliki tahun yang sama. 
model_year_group = df.groupby(['model'])['model_year'].agg(lambda x: x.value_counts().index[0]).to_dict()
model_year_group
````

````python
# Nilai yang hilang pada kolom model_year diisi dengan tahun jenis mobil yang sama mengikuti kolom model
df['model_year'] = df['model_year'].fillna(df['model'].map(model_year_group))
````

````python
# Pengecekan nilai yang hilang pada kolom cylinders
df['cylinders'].unique()
````

````python
cylinders_group = df.groupby(['model'])['cylinders'].agg(lambda x: x.value_counts().index[0]).to_dict()
````

````python
# Nilai yang hilang pada kolom cylinders diisi dengan tahun jenis mobil yang sama mengikuti kolom model karena mobil dengan tipe yang sama kemungkinan besar memiliki jumlah silinder yang sama.
df['cylinders'] = df['cylinders'].fillna(df['model'].map(cylinders_group))
````

## Mempelajari dan Menangani Outlier <a id=8></a>



<div class="alert alert-info">
<b>Chamdani's info v.1</b> <a class="tocSkip"></a>

Outlier merupakan nilai data yang tak biasa dari suatu kolom. Dia terkadang bisa lebih kecil atau lebih besar dari kecenderungan nilai dari kolom tersebut. Lihatlah histogram yang kamu buat diatas, range nilai pada sumbu x terlihat sangat jauh. Hal ini memperlihatkan bahwa kemungkinan percilan atau outlier berada pada rentang nilai diatas nilai Q3 dan upper limit dari boxplot. Dan itu terbukti dalam boxplot yang kamu tampilkan diatas. 
    
**Perhatikan gambaran berikut**


</div>

````python
# Tentukan batas bawah outlier 
````

````python
# Tentukan batas atas outlier 
````

````python
list_of_column = ['price', 'age_car', 'odometer', 'cylinders', 'condition']

def outlier(column_name=''):
    q1 = df[column_name].quantile(0.25)
    q3 = df[column_name].quantile(0.75)
    iqr = q3 - q1
    lower = q1 - 1.5 * iqr
    upper = q3 + 1.5 * iqr
    if q1 - 1.5 * iqr < 0:
        return (0, upper)
    else:
        return (lower, upper)
    
df_filter_outlier = df.copy()
````

````python
# Cek outlier
price_outlier = outlier(column_name='price')
price_outlier
````

````python
age_car_outlier = outlier(column_name='age_car')
age_car_outlier
````

````python
odometer_outlier = outlier(column_name='odometer')
odometer_outlier
````

````python
cylinders_outlier = outlier(column_name='cylinders')
cylinders_outlier
````

````python
condition_outlier = outlier(column_name='condition')
condition_outlier
````

````python
for column in list_of_column:
    print('start cleaning outlier on ', column)
    lower, upper = outlier(column_name=column)
    #lower = outlier([0])
    #upper = outlier[1]
    df_filter_outlier = df_filter_outlier[(df_filter_outlier[column] > lower) & (df_filter_outlier[column] < upper )]
    print('cleaning outlier on ', column, ' was finished!')
````

````python
# Boxplot untuk variabel kategorik
def boxplot(df, a, b, title=None):
    df_list = df[b].value_counts().reset_index().query(f'{b} >= 50')['index'].to_list()
    df_filter_50 = df.query(f'{b} in @df_list')
    df_filter_50.boxplot(column=a, by=b, figsize=(10,5))
    plt.title(title)
    plt.tight_layout()
    plt.show()
````

````python
boxplot(suv, 'price', 'transmission', title='SUV')
boxplot(sedan, 'price', 'transmission', title='Sedan')
````
