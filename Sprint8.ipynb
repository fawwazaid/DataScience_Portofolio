## Training Model

Training akan dilakukan dengan 3 model yaitu Decision Tree, Random Forest, dan Logistic Regression.

### Decision Tree

```python
best_score_dt = 0
best_depth_dt = 0
for depth in range(1,41):
    model = DecisionTreeClassifier(random_state=12345, max_depth=depth, class_weight='balanced')
    model.fit(X_train, y_train)
    probabilities_valid = model.predict_proba(X_valid)
    probabilities_one_valid = probabilities_valid[:, 1]
    fpr, tpr, thresholds = roc_curve(y_valid, probabilities_one_valid)
    y_pred = model.predict(X_valid)
    train_pred = model.predict(X_train)
    train_acc = accuracy_score(y_train, train_pred)
    score = accuracy_score(y_valid, y_pred)
    report_valid = classification_report(y_valid, y_pred, output_dict=True) 
    
    if score > best_score_dt:
        best_score_dt = score
        best_depth_dt = depth
        accuracy_diff = score - train_acc
        best_train = train_acc
        
print(train_acc)
print(accuracy_diff)
print(f'Akurasi model Desicision Tree terbbaik pada validation set (max_depth = {best_depth_dt}) : {best_score_dt}')
print(f'F1 score:{f1_score(y_valid, y_pred)}')
display(pd.DataFrame(confusion_matrix(y_valid, y_pred)))
display(pd.DataFrame(report_valid))
plt.figure()
plt.plot(fpr, tpr)
plt.plot([0, 1], [0, 1], linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.0])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Kurva ROC')
plt.show()
```

Hasil yang didapatkan model Decision Tree cukup bagus, hanya saja training akurasinya terlalu fit sehingga kemungkinan besar modelnya akan overfitting. Skor f1 yang didapatkan masih dibawah 0.59.

### Random Forest

```python
best_score_clf = 0
best_est_clf = 0

for est in range(10,306,5):
    model = RandomForestClassifier(random_state=12345, n_estimators=est, max_depth=3, class_weight='balanced')
    model.fit(X_train, y_train)
    probabilities_valid = model.predict_proba(X_valid)
    probabilities_one_valid = probabilities_valid[:, 1]
    fpr, tpr, thresholds = roc_curve(y_valid, probabilities_one_valid)
    y_pred = model.predict(X_valid)
    train_pred = model.predict(X_train)
    train_acc = accuracy_score(y_train, train_pred)
    report_valid = classification_report(y_valid, y_pred, output_dict=True) 
    score = accuracy_score(y_valid, y_pred)
    if score > best_score_clf:
        best_score_clf = score
        best_est_clf = est
        accuracy_diff = score - train_acc
        best_train = train_acc
        
print(train_acc)
print(accuracy_diff)
print(f'Akurasi model Random Forest terbaik pada validation set (n_estimator = {best_est_clf}) : {best_score_clf}')
print(f'F1 score:{f1_score(y_valid, y_pred)}')
display(pd.DataFrame(confusion_matrix(y_valid, y_pred)))
display(pd.DataFrame(report_valid))
plt.figure()
plt.plot(fpr, tpr)
plt.plot([0, 1], [0, 1], linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.0])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Kurva ROC')
plt.show()
```

Disini kita mencoba model random forest dengan max depth dari model decision tree sebelumnya. Hasil yang didapatkan cukup bagus namun akurasinya tidak terlalu tinggi. Selanjutnya kita akan mencoba untuk sedikit merubah hyperparameternya.

### Upsampling

Pertama kita akan mencoba teknik upsampling pada train dataset.

```python
def upsample(features, target, repeat):
    features_zeros = features[target == 0]
    features_ones = features[target == 1]
    target_zeros = target[target == 0]
    target_ones = target[target == 1]

    features_upsampled = pd.concat([features_zeros] + [features_ones] * repeat)
    target_upsampled = pd.concat([target_zeros] + [target_ones] * repeat)

    features_upsampled, target_upsampled = shuffle(
        features_upsampled, target_upsampled, random_state=12345
    )
    
    return features_upsampled, target_upsampled
```

### Downsampling

Disini kita akan mencoba teknik downsampling pada train dataset.

```python
def downsample(features, target, fraction):
    features_zeros = features[target == 0]
    features_ones = features[target == 1]
    target_zeros = target[target == 0]
    target_ones = target[target == 1]

    features_downsampled = pd.concat(
        [features_zeros.sample(frac=fraction, random_state=12345)]
        + [features_ones]
    )
    target_downsampled = pd.concat(
        [target_zeros.sample(frac=fraction, random_state=12345)]
        + [target_ones]
    )

    features_downsampled, target_downsampled = shuffle(
        features_downsampled, target_downsampled, random_state=12345
    )

    return features_downsampled, target_downsampled
```
